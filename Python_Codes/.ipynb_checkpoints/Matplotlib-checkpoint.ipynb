## 1. Matplotlib
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

## 1.1. Matplotlib Pyplot
Most of the Matplotlib utilities lies under the <code style= "color:red">pyplot</code> submodule, and are usually imported under the <code style= "color:red">plt</code> alias.

```python
# Draw a line in a diagram from position (0,0) to position (6,250):
import numpy as np
import matplotlib.pyplot as plt

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])


plt.plot(xpoints, ypoints, marker = 'x')
plt.show()
```

## Markers
<br>There are different markers one can use. 

```python
# Markers
import pandas as pd

data_1 = {'Markers': ['o', '$*$', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', 
                       '>', '1', '2', '3', '4', '|', '_'], 
           'Description': ['Circle', 'Star' ,'Point' ,'Pixel' ,'X' ,'X (filled)' ,'Plus' 
                           ,'Plus (filled)', 'Square' ,'Diamond' ,'Diamond (thin)' ,'Pentagon' 
                           ,'Hexagon' ,'Hexagon' ,'Triangle Down' ,'Triangle Up' ,'Triangle Left', 
                           'Triangle Right' ,'Tri Down' ,'Tri Up' ,'Tri Left' ,'Tri Right' ,'Vline', 'Hline']}

data_1_pandas = pd.DataFrame(data_1)
display(data_1_pandas)
```

## Line Style
<br>We can choose any of these styles.

```python
data_2 = {'Style':['solid (defauld)', 'dotted', 'dashed', 'dashdot', 'none'], 'Or':['-', ':', '--', '-.', 'or']}
data_2_pandas = pd.DataFrame(data_2)
display(data_2_pandas)
```

```python

plt.plot(xpoints, ypoints, linestyle = 'dotted')
plt.show()
```

**Shorter Syntax:**
<br> The line style can be written in a shorter syntax:
<br> <code style= "color:red">linestyle</code> can be written as <code style= "color:red">ls</code>.
<br> <code style= "color:red">dotted</code> can be written as <code style= "color:red">.</code>.
<br> <code style= "color:red">dashed</code> can bewritten as <code style= "color:red">-</code>.

```python
plt.plot(xpoints, ypoints, ls = '--')
plt.show()
```

## Line Color
<br> We can use the keyword argument <code style= "color:red">color</code> or the shorter <code style= "color:red">c</code> to set the color of the line.
<br>
<br> We can use any color from 140 different colors either calling by [color name](https://www.w3schools.com/colors/colors_names.asp) or by [hexadecimal color value](https://www.w3schools.com/colors/colors_hexadecimal.asp). 

```python
plt.plot(ypoints, ls='--', color = 'r')
# or
# plt.plot(ypoints, ls='--', c = 'r')
plt.show()
```

## Line Width
<br> You can use the keyword argument <code style= "color:red">linewidth</code> or the shorter <code style= "color:red">lw</code> to change the width of the line. The value is a floating number, in points.

```python
plt.plot(ypoints, ls = '-', c = 'b', linewidth = '5.5')
plt.show()
```

## Multiple Lines

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

# plt.plot(x1, y1, x2, y2)
# or
plt.plot(x1, y1, ls = '-', c = 'r', lw = 1.0)
plt.plot(x2, y2, ls = ':', c = 'b', lw = .5)
plt.show()
```

## Matplotlib Labels and Title
<br> With Pyplot, you can use the <code style= "color:red">xlabel()</code> and <code style= "color:red">ylabel()</code> functions to set a label for the x- and y-axis.

```python
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.plot(x1, y1, ls = '-', c = 'r', lw = 1.0)
plt.show
```

## Create a Title for a Plot
With Pyplot, you can use the <code style= "color:red">title()</code> function to set a title for the plot.

```python
plt.title('This is the plot title')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.plot(x2, y2, ls = '-', c = 'r', lw = 1.0)
plt.grid()
plt.show
```

**Font siz:** We can also choose the font size by adding the following lines in the code.
<br> <code style= "color:red">font1 = {'family':'serif','color':'blue','size':20}</code>.
<br> <code style= "color:red">font2 = {'family':'serif','color':'darkred','size':15}</code>.
<br>
<br> And also we can change the position of the title by adding the following line in the code.
<br> <code style= "color:red">plt.title("Sports Watch Data", loc = 'left')</code>


## Add Grid Lines to a Plot
To add **grid lines** to a plot we can add <code style= "color:red">plt.grid()</code>. See example above.
<br> We can also specify which grid lines to show by adding the followin line in the code.
<br> <code style= "color:red">plt.grid(axis = 'x')</code>
<br> or
<br> <code style= "color:red">plt.grid(axis = 'y')</code>
<br>
<br> **Set Line Properties for the Grid**
<br> You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).
<br> <code style= "color:red">plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)</code>

## 1.2. Matplotlib Subplot
With the <code style= "color:red">subplot()</code> function you can draw multiple plots in one figure.

```python
#plot 1:
x_2 = np.array([0, 1, 2, 3])
y_2 = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # the figure has 1 row and 2 columns, and this plot is the first plot.
plt.plot(x_2,y_2)

#plot 2:
x_3 = np.array([0, 1, 2, 3])
y_3 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # the figure has 1 row and 2 columns, and this plot is the second plot.
plt.plot(x_3,y_3)

plt.show()
```

## 1.3 My Matplotlib Script

```python
# Loading the data 
data100=np.loadtxt('./280_60(100).dat',usecols=(0,1,2,3,4,5,6))
calc100=np.loadtxt('./../RFG_electrons/12C/second-run/fort.100')
# Plotting the data
fig, axs = plt.subplots(1,figsize=(8,5))
axs.plot(calc100[:,0],calc100[:,1], ls='-',c='gray',label='RFG Calculation')
axs.plot(data100[:,4],data100[:,5]/10**(3),ls='x',c='red',label='Experimental data')
axs.set_title('$E_i=280$ MeV, ${\\theta}=60^\circ$',fontsize=14,style='normal')
axs.set_xlim([0.0,0.15])
axs.set_xlabel('${\omega}$ [GeV]')
axs.set_ylabel('d${\sigma}$/d${\rho}$dE [nb/sr/GeV]')
axs.legend()
plt.grid(true)
plt.savefig('280_60.pdf')
```
